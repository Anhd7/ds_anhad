```python
import pandas as pd
```

```python
sentiment = pd.read_csv("csv_files/fear_greed_index.csv")
trades = pd.read_csv("csv_files/historical_data.csv")
```

```python
sentiment['date'] = pd.to_datetime(sentiment['date']).dt.date
```

```python
#sentiment['date'].isna().sum()
```

```python
trades['Timestamp IST'] = pd.to_datetime(trades['Timestamp IST'], format='%d-%m-%Y %H:%M',dayfirst=True)
trades['date'] = trades['Timestamp IST'].dt.date
```

```python
#trades['date'].isna().sum()
```

```python
trades['Closed PnL'].astype(str).head(10)
```

```python
trades['Closed PnL'] = (trades['Closed PnL'].astype(str).str.replace(',', '', regex=False).str.strip())
```

```python
trades['Closed PnL'] = pd.to_numeric(trades['Closed PnL'],errors='coerce')
```

```python
trades['Closed PnL'].describe()
```

```python
trades['Size USD'] = (trades['Size USD'].astype(str).str.replace(',', '', regex=False).str.strip())
```

```python
trades['Size USD'] = pd.to_numeric(trades['Size USD'],errors='coerce')
```

```python
trades['Size USD'].describe()
```

```python
trades = trades.dropna(subset=['Closed PnL', 'Size USD'])
```

```python
sentiment_map = {'Extreme Fear': -2,'Fear': -1,'Neutral': 0,'Greed': 1,'Extreme Greed': 2}
sentiment['sentiment_score'] = sentiment['classification'].map(sentiment_map)
```

```python
sentiment[['classification', 'sentiment_score', 'value']].drop_duplicates().sort_values('sentiment_score')
```

```python
sentiment['date'].min(), sentiment['date'].max()
```

```python
trades['date'].min(), trades['date'].max()
```

```python
merged = trades.merge(sentiment[['date', 'classification', 'sentiment_score', 'value']],on='date',how='inner')
```

```python
merged.shape
```

```python
merged['Closed PnL'].describe()
```

```python
merged.sort_values(by='Closed PnL', ascending=False).head(10)
```

```python
merged['classification'].value_counts()
```

```python
merged[['Closed PnL', 'Size USD']].isna().sum()
```

```python
merged['profit'] = (merged['Closed PnL'] > 0).astype(int)
merged['pnl_per_usd'] = merged['Closed PnL'] / merged['Size USD']
```

```python
merged[['Closed PnL', 'Size USD', 'pnl_per_usd']].describe()
```

```python
analysis_df = merged[['date','classification','sentiment_score','value','Closed PnL','Size USD','profit','pnl_per_usd']].copy()
```

```python
summary = analysis_df.groupby('classification').agg(trades=('Closed PnL', 'count'),profit_rate=('profit', 'mean'),avg_trade_size=('Size USD', 'mean'),avg_pnl=('Closed PnL', 'mean'),median_pnl=('Closed PnL', 'median'),efficiency=('pnl_per_usd', 'mean')).sort_values('efficiency', ascending=False)
```
